```python
### YOUR CODE HERE ###
import pandas as pd
import datetime
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib
from pandas.plotting import register_matplotlib_converters
df_course_cont=pd.read_csv('course_contents.csv')
df_courses=pd.read_csv('courses.csv')
df_progress_ph=pd.read_csv('progress_phases.csv')
df_progresses=pd.read_csv('progresses.csv')
```

```python
#Создание общего датасета
course_full=pd.merge(left=df_course_cont, right=df_courses, left_on='course_id', right_on='id')
df_progr_stud=pd.merge(left=df_progresses, right=course_full, on='course_id')
df_progr_stud=df_progr_stud.rename(columns={'id_x': 'progress_id'})
df_full=pd.merge(left=df_progr_stud, right=df_progress_ph, on=['progress_id','module_number','lesson_number'])
df_full=df_full.drop('id_y', axis=1)
```

```python
# общее количество курсов в датасете
df_full['title'].nunique()
```

```python
# количество модулей на каждом курсе
df_full[['title','module_title']].groupby('title').agg({"module_title": 'nunique'})
```

```python
#количество уроков в каждом модуле на каждом курсе
lesson_num=df_full[['title','module_title','lesson_title']].groupby(['title','module_title']).agg({'lesson_title': 'nunique'})
lesson_num
```

```python
#медианное количество уроков в модуле на каждом курсе

np.median(lesson_num['lesson_title'])
```

```python
#количество учеников на каждом курсе

stud_course_num=df_full[['title','student_id']].groupby('title').agg({"student_id": 'nunique'})
stud_course_num
```

```python
#минимальный, максимальный, средний, медианный возраст студентов

pd_students=pd.read_csv('students.csv')
pd_students.dropna(subset=['birthday'], inplace=True)
f=lambda x: pd.to_datetime(x, errors = 'coerce')
z = lambda x: math.floor(x)
pd_students['birthday']=pd_students.loc[:,'birthday'].apply(f)
pd_students['ages']=(pd.to_datetime(datetime.datetime.now().strftime('%Y-%m-%d'))-pd_students['birthday'])/np.timedelta64(1,'Y')
pd_students.dropna(subset=['birthday'], inplace=True)
pd_students['ages']=pd_students.loc[:,'ages'].apply(z)
pd_students['ages'].agg(['min', 'max','mean', 'median'])
```

```python
#минимальный, максимальный, средний, медианный возраст студентов на каждом курсе

stud_course_id=pd.merge(left=df_progresses, right=pd_students, left_on='student_id', right_on='id')[['id_x','ages','course_id']]
stud_course_id
stud_course=pd.merge(left=stud_course_id, right=df_courses, left_on='course_id', right_on='id')[['id_x','ages','title']]
st=stud_course[['title','ages']].groupby('title').agg(['min', 'max','mean', 'median'])
st
```

```python
#График распределения студентов по каждому курсу
def foo(x):
    %matplotlib inline
    register_matplotlib_converters()
    fig, subplot = plt.subplots(figsize=(15,10))
    y_ticks=np.array(x['student_id'])
    x_ticks=np.array(x.index)
    subplot.bar(x_ticks,y_ticks)
    subplot.set_title('Число студентов на каждом курсе', fontsize=15)
    subplot.set_xlabel('Название курса', fontsize=15)
    plt.grid(linewidth=0.5, alpha=0.3)
    plt.xticks(x_ticks, rotation='vertical')
    plt.show()
foo(stud_course_num)
```

```python
#График распределения студентов по каждому курсу с медианой

df={'Course': np.array(stud_course_num.index),'Number': np.array(stud_course_num['student_id'])}
stud1=pd.DataFrame(df)
stud1.sort_values('Number', inplace=True)
%matplotlib inline
register_matplotlib_converters()
fig, subplot = plt.subplots(figsize=(15,10), sharex=True)
y_ticks=stud1.loc[:,'Number']
x_ticks=stud1.loc[:,'Course']
median=np.median(stud1.loc[:,'Number'])
courses=subplot.barh(x_ticks,y_ticks, color =['#FF0000','#00FF00','#0000FF','#FFFF00',
                                      '#00FFFF','#FF00FF','#C0C0C0','#808080',
                                      '#800000','#808000','#008000','#800080',
                                      '#008080','#000080','#2F4F4F'])
subplot.set_title('Число студентов на каждом курсе', fontsize=15)
subplot.set_xlabel('Кол-во студентов', fontsize=15)
subplot.set_frame_on(False)
plt.grid(linewidth=0.5, alpha=0.1)
plt.axvline(median,linestyle='-', color='crimson', linewidth=5)
plt.text(median, 1, 'median', fontsize=20, horizontalalignment='left', rotation='vertical')
subplot.legend(courses, stud1.loc[:,'Course'],
          title="Course",
          loc=4)

plt.show()
```

```python
### YOUR CODE HERE ###
### Прирост студентов
student_hw=df_full[df_full['is_homework']==True][['title','student_id', 'start_date']]
student_hw['start_date']=pd.to_datetime(student_hw['start_date'])
student_hw['start_date']=student_hw['start_date'].map(lambda x: pd.to_datetime(x.strftime('%m %Y')))
student_hw.sort_values('start_date', inplace=True)
student_hw_groups=student_hw.groupby(['title', 'start_date']).agg({"student_id": 'nunique'}).sort_values(['title', 'start_date', 'student_id'])
student_hw_groups['difference']=student_hw_groups['student_id'].diff()
student_hw_groups.reset_index(inplace=True)
for i in range(len(student_hw_groups['title'])):
    try:
        if student_hw_groups.loc[i,'title']!=student_hw_groups.loc[i-1,'title']:
            student_hw_groups.loc[i,'difference']=0
    except Exception:
        pass
student_hw_groups.fillna(0, inplace=True)
student_hw_groups
```

```python
### 15 графиков с приростом
groups_list=student_hw_groups['title'].unique()
for i in groups_list:
    %matplotlib inline
    register_matplotlib_converters()
    fig, subplot = plt.subplots(figsize=(15,3), sharex=True)
    graph=student_hw_groups[student_hw_groups['title'] == i]
    y_ticks=graph.loc[:,'difference']
    x_ticks=graph.loc[:,'start_date'].astype(str)
    subplot.plot(x_ticks,y_ticks)
    subplot.set_title(i, fontsize=15)
    plt.xticks(x_ticks, rotation='vertical')
    plt.show()
```

```python
#График с 15 линиями
%matplotlib inline
fig, subplot = plt.subplots(figsize=(15,10), sharex=True)
colors =['#FF0000','#00FF00','#0000FF','#FFFF00',
                                      '#00FFFF','#FF00FF','#C0C0C0','#808080',
                                      '#800000','#808000','#008000','#800080',
                                      '#008080','#000080','#2F4F4F']
num=0
for i in groups_list:
    graph=student_hw_groups[student_hw_groups['title'] == i]
    y_ticks=graph.loc[:,'difference']
    x_ticks=graph.loc[:,'start_date']
    subplot.plot(x_ticks,y_ticks, color=colors[num])
    num+=1
plt.xticks(rotation='vertical')
plt.yticks(np.arange(-100,200, 10))
subplot.xaxis.set_major_locator(matplotlib.dates.MonthLocator(interval=2))
subplot.set_title('Прирост студентов на каждом курсе', fontsize=15)
subplot.set_ylabel('Прирост', fontsize=15)
subplot.legend(groups_list, title="Course", loc=2)
subplot.set_frame_on(False)
subplot.grid(linewidth=0.5, alpha=0.3)
plt.show()
```

```python
#Кол-во прогрессов
import datetime as dt
df_progress=df_full[df_full['is_homework']==True][['title', 'progress_id', 'start_date', 'finish_date','status']]
df_progress['month']=pd.to_datetime(df_progress['start_date']).map(lambda x: pd.to_datetime(x.strftime('%m %Y')))
df_progress['start_date']=pd.to_datetime(df_progress['start_date']).map(lambda x: pd.to_datetime(x.strftime('%m %Y'))).dt.to_period('M')
df_progress['finish_date']=pd.to_datetime(df_progress['finish_date']).map(lambda x: pd.to_datetime(x.strftime('%m %Y'))).dt.to_period('M')
df_progress['finish_date'].fillna(df_progress['start_date'], inplace=True)
df_progress['months_progress']=(df_progress['finish_date'].astype(int)-df_progress['start_date'].astype(int))+1
progress_numb=df_progress[['title','start_date','month','months_progress']].groupby(['title','start_date','month']).agg('sum')
progress_numb.sort_values('start_date', inplace=True)
progress_numb.reset_index(inplace=True)
progress_numb
```

```python
### 15 графиков с прогрессами
groups_progress=progress_numb['title'].unique()
%matplotlib inline
register_matplotlib_converters()
for i in groups_progress:
    fig, subplot = plt.subplots(figsize=(15,3), sharex=True)
    progresses=progress_numb[progress_numb['title'] == i]
    y_ticks=progresses.loc[:,'months_progress']
    x_ticks=progresses.loc[:,'month'].astype(str)
    subplot.plot(x_ticks,y_ticks)
    subplot.set_title(i, fontsize=15)
    plt.xticks(x_ticks, rotation='vertical')
    plt.show()
```

```python
#График с 15 линиями
%matplotlib inline
fig, subplot = plt.subplots(figsize=(15,10), sharex=True)
colors =['#FF0000','#00FF00','#0000FF','#FFFF00',
                                      '#00FFFF','#FF00FF','#C0C0C0','#808080',
                                      '#800000','#808000','#008000','#800080',
                                      '#008080','#000080','#2F4F4F']
num=0
for i in groups_list:
    graph=progress_numb[progress_numb['title'] == i]
    y_ticks=graph.loc[:,'months_progress']
    x_ticks=graph.loc[:,'month']
    groups_line=subplot.plot(x_ticks,y_ticks, color=colors[num])
    num+=1
subplot.set_title('Прогресс на каждом курсе', fontsize=15)
subplot.set_ylabel('Прогресс', fontsize=15)
plt.xticks(rotation='vertical', fontsize=15)
subplot.xaxis.set_major_locator(matplotlib.dates.MonthLocator(interval=2))
plt.yticks(np.arange(0,2500, 100))
subplot.legend(groups_list, title="Course", loc=2)
subplot.set_frame_on(False)
subplot.grid(linewidth=0.5, alpha=0.3)
plt.show()
```

```python
### Расчет конверсии перехода студентов из одного модуля в другой

df_conversion=df_full.loc[:,['title', 'module_title', 'module_number','lesson_number',
                    'status', 'start_date', 'finish_date']][(df_full['is_homework']==True)&((df_full['status']=='done')|(df_full['status']=='start'))]
df_conversion['start_date']=pd.to_datetime(df_conversion['start_date']).map(lambda x: pd.to_datetime(x.strftime('%m %d %Y')))
df_conversion['finish_date']=pd.to_datetime(df_conversion['finish_date']).map(lambda x: pd.to_datetime(x.strftime('%m %d %Y')))
df_conversion_start=df_conversion.groupby(['title','module_title', 'module_number','lesson_number',]).agg({'start_date':'count'}).reset_index()
df_conversion_finish=df_conversion.groupby(['title','module_title', 'module_number','lesson_number',]).agg({'finish_date':'count'}).reset_index()
df_conversion_start_min=df_conversion_start.groupby(['title','module_title', 'module_number']).agg({'lesson_number':'min'}).reset_index()
df_conversion_start_max=df_conversion_start.groupby(['title','module_title', 'module_number']).agg({'lesson_number':'max'}).reset_index()
df_merged_conversion=pd.merge(pd.merge(df_conversion_start, df_conversion_start_min, on=['title', 'module_number', 'lesson_number']),
pd.merge(df_conversion_finish, df_conversion_start_max, on=['title', 'module_number', 'lesson_number']), 
on=['title', 'module_number'])[['title','module_title_x_x','module_number','start_date', 'finish_date']].sort_values(['title','module_number'])
df_merged_conversion.reset_index(inplace=True)
df_merged_conversion.drop('index', axis=1)
df_merged_conversion['conversion']=0
for i in range(1, len(df_merged_conversion)):   
    if df_merged_conversion.loc[i,'title']==df_merged_conversion.loc[i-1,'title']:
        df_merged_conversion.loc[i,'conversion']=df_merged_conversion.loc[i,'start_date']/df_merged_conversion.loc[i-1,'finish_date']
df_merged_conversion.replace([np.inf, -np.inf], np.nan, inplace=True)
df_merged_conversion.fillna(1, inplace=True)
df_field=df_full[df_full['is_homework']==True][['title','module_title', 'field']].drop_duplicates()
df_merged_conversion
```

```python
###bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе

%matplotlib inline
register_matplotlib_converters()
fig, subplot = plt.subplots(figsize=(20,15), sharex=True)
y_ticks=df_merged_conversion.loc[:,'conversion']
x_ticks=df_merged_conversion.loc[:,'module_title_x_x']
subplot.bar(x_ticks,y_ticks)
plt.xticks(rotation='vertical', fontsize=8)
subplot.grid(linewidth=0.5, alpha=0.3)
plt.show()
```

```python
###горизонтальный bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. 15 графиков

df_merged_conversion['module_title_number']=df_merged_conversion['module_number'].astype(str)+'. '+df_merged_conversion['module_title_x_x']
course_title=df_merged_conversion['title'].unique()
colors=['#FF0000','#00FF00','#0000FF','#FFFF00',
                                      '#00FFFF','#FF00FF','#C0C0C0','#808080',
                                      '#800000','#808000','#008000','#800080',
                                      '#008080','#000080','#2F4F4F']
num=0
for i in course_title:
    %matplotlib inline
    register_matplotlib_converters()
    fig, subplot = plt.subplots(figsize=(20,5), sharex=True)
    course=df_merged_conversion[df_merged_conversion['title'] == i]
    y_ticks=course.loc[:,'conversion']
    x_ticks=course.loc[:,'module_title_number']
    modules=subplot.barh(x_ticks,y_ticks, color=colors[num])
    subplot.set_title(i, fontsize=20)
    median=np.median(y_ticks)
    plt.axvline(median,linestyle='-', color='pink', linewidth=2)
    subplot.set_frame_on(False)
    subplot.grid(linewidth=0.5, alpha=0.3)
    field=df_field[df_field['title']==i]['field'].unique()
    subplot.legend(modules, field,
           title="field", loc=4)
    num+=1
    plt.show()
```
